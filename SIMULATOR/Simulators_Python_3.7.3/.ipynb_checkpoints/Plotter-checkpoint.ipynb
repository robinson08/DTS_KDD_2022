# Import necessary libraries

```python
#Pandas and plotting tools
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
```

# Import the RESULTS as a Dataframe

```python
# Read the dataframe
results_batches = pd.read_csv("./Results/Batches/Batch_times_V3.csv")
results_epochs = pd.read_csv("./Results/Epochs/Epoch_times.csv")
```

# Plotting

## 1. Plot Batch Times, Real VS Simulated

### Version 1: Linear Plot

```python
# Plot batches

# Set plot dimensions
plt.figure(figsize = (15,7.5))

# Set the name of the X axis
plt.xlabel("Batch #", size = 15)

# Set the name of the Y axis
plt.ylabel("Batch Training Time [s]", size = 15)

# Set the title of the plot
plt.title("Real Batch Training Time VS Simulated Batch Training Time", size = 20)

# Plot each of the algorithms' TRAINING TIMES VS x_variable[i]
# 'Measured PL' would refer to the time for ONE batch, so 'Measured PL Equivalent' refers to 
# the time of training PL with the same amount of data as all the other architectures. Remember
# PL only has one source, so if the other architectures have 3 clients, they have 3 sources.
# Hence, if the others have three sources with 100 batches each, PL must have 3 batches of 100
# in one source for the comparison to be equivalent between the architectures
plt.scatter(results_batches['Batch #'], results_batches['Measured PL Equivalent'], label = "Real PL Time", 
         marker = "o", color = 'black')
plt.plot(results_batches['Batch #'], results_batches['Predicted PL Equivalent'], label = "Simulated PL Time", 
         color = 'black', linestyle = 'dashed')
# plt.scatter(results_batches['Batch #'], results_batches['Predicted PL Equivalent'], label = "Predicted PL Time", 
#          marker = "x", color = 'black')

# plt.scatter(results_batches['Batch #'], results_batches['Measured PL'], label = "Measured PL Time", 
#          marker = "o", color = 'black')
# plt.scatter(results_batches['Batch #'], results_batches['Predicted PL'], label = "Predicted PL Time", 
#          marker = "x", color = 'black')

plt.scatter(results_batches['Batch #'], results_batches['Measured FL'], label = "Real FL Time", 
         marker = "o", color = 'blue')
plt.plot(results_batches['Batch #'], results_batches['Predicted FL'], label = "Simulated FL Time", 
         color = 'blue', linestyle = 'dashed')
# plt.scatter(results_batches['Batch #'], results_batches['Predicted FL'], label = "Predicted FL Time", 
#          marker = "x", color = 'blue')

plt.scatter(results_batches['Batch #'], results_batches['Measured SL'], label = "Real SL Time", 
         marker = "o", color = 'red')
plt.plot(results_batches['Batch #'], results_batches['Predicted SL'], label = "Simulated SL Time", 
         color = 'red', linestyle = 'dashed')
# plt.scatter(results_batches['Batch #'], results_batches['Predicted SL'], label = "Predicted SL Time", 
#          marker = "x", color = 'red')

plt.scatter(results_batches['Batch #'], results_batches['Measured PSL'], label = "Real PSL Time", 
         marker = "o", color = 'green')
plt.plot(results_batches['Batch #'], results_batches['Predicted PSL'], label = "Simulated PSL Time", 
         color = 'green', linestyle = 'dashed')
# plt.scatter(results_batches['Batch #'], results_batches['Predicted PSL'], label = "Predicted PSL Time", 
#          marker = "x", color = 'green')

plt.scatter(results_batches['Batch #'], results_batches['Measured FSL'], label = "Real FSL Time", 
         marker = "o", color = 'yellow')
plt.plot(results_batches['Batch #'], results_batches['Predicted FSL'], label = "Simulated FSL Time", 
         color = 'yellow', linestyle = 'dashed')
# plt.scatter(results_batches['Batch #'], results_batches['Predicted FSL'], label = "Predicted FSL Time", 
#          marker = "x", color = 'yellow')

# Define your x ticks
x_ticks = results_batches['Batch #']

# Define your x and y axis limits
plt.ylim(0, 0.8)

# Set the legend and show the plot
plt.legend()

# Save your image
file_name = "BATCHES_Real_VS_Simulator_V2"
plt.savefig(fname = './Results/Batches/' + file_name + '.png')

# Show your image
plt.show()
```

### Version 2: Boxplot

```python
# Plot batches

# Set plot dimensions
plt.figure(figsize = (15,7.5))

# Set the name of the X axis
# plt.xlabel("Distributed Training Architecture", size = 15)
plt.xlabel("Distributed Training Architecture", size = 20)

# Set the name of the Y axis
# plt.ylabel("Batch Training Time [s]", size = 15)
plt.ylabel("Batch Training Time [s]", size = 20)

# Set the title of the plot
# plt.title("Real Batch Training Time VS Simulated Batch Training Time", size = 20)

# Plot each of the algorithms' TRAINING TIMES VS x_variable[i]
# 'Measured PL' would refer to the time for ONE batch, so 'Measured PL Equivalent' refers to 
# the time of training PL with the same amount of data as all the other architectures. Remember
# PL only has one source, so if the other architectures have 3 clients, they have 3 sources.
# Hence, if the others have three sources with 100 batches each, PL must have 3 batches of 100
# in one source for the comparison to be equivalent between the architectures

# Using Pandas boxplot() because plt.boxplot() does not deal nicely with NaN values!
# dataframe.boxplot() will take any info I gave to plt.whatever and use it for its own info!
# The following code will also put "both" boxplots (predicted and real) on the same x-tick!

# First make the boxplot, but return the axes... 
color_testbed = 'blue'
ax = results_batches.boxplot(column=['Measured PL Equivalent','Measured FL','Measured SL', 
                                     'Measured PSL','Measured FSL'], 
                                     grid = False, 
                                     fontsize = 15,
                                     showfliers = False,
                                     color = color_testbed,
                                     return_type = 'axes')
#...Then tell the next boxplot to use the same axes!
color_predicted = 'red'
results_batches.boxplot(column=['Predicted PL Equivalent','Predicted FL','Predicted SL', 
                                     'Predicted PSL','Predicted FSL'], 
                                     grid = False, 
                                     fontsize = 15,
                                     medianprops=dict(linestyle='-', linewidth=2),
                                     showfliers = False,
                                     color = color_predicted,
                                     ax = ax)

plt.xticks([1, 2, 3, 4, 5], ['PL', 'FL', 'SL', 'PSL', 'FSL'])

# Define your x ticks
x_ticks = results_batches['Batch #']

# Define your x and y axis limits
plt.ylim(ymin = 0, ymax = 0.5)

# Set our own custom legend
red_patch = mpatches.Patch(color='red', label='Simulations')
# blue_patch = mpatches.Patch(color='green', label='Prototype Implementation')
blue_patch = mpatches.Patch(color='blue', label='Testbed')
plt.rcParams["legend.fontsize"] = 15
plt.legend(handles=[red_patch, blue_patch])

# Save your image
file_name = "BATCHES_Real_VS_Simulator_Box_V2"
plt.savefig(fname = './Results/Batches/' + file_name + '.png')

# Show your image
plt.show()
```

## 2. Plot Epoch Times, Real VS Simulated

```python
# Plot epochs

# Set plot dimensions
plt.figure(figsize = (15,7.5))

# Set the name of the X axis
plt.xlabel("Epoch #", size = 15)

# Set the name of the Y axis
plt.ylabel("Epoch Training Time [s]", size = 15)

# Set the title of the plot
plt.title("Real Epoch Training Time VS Simulated Epoch Training Time", size = 20)

# Plot each of the algorithms' TRAINING TIMES VS x_variable[i]
plt.scatter(results_epochs['Epoch #'], results_epochs['Measured PL'], label = "Real PL Time", 
         marker = "o", color = 'black')
plt.plot(results_epochs['Epoch #'], results_epochs['Predicted PL'], label = "Simulated PL Time", 
         color = 'black', linestyle = 'dashed')
# plt.scatter(results_epochs['Epoch #'], results_epochs['Predicted PL'], label = "Predicted PL Time", 
#          marker = "x", color = 'black')


plt.scatter(results_epochs['Epoch #'], results_epochs['Measured FL'], label = "Real FL Time", 
         marker = "o", color = 'blue')
plt.plot(results_epochs['Epoch #'], results_epochs['Predicted FL'], label = "Simulated FL Time", 
         color = 'blue', linestyle = 'dashed')
# plt.scatter(results_epochs['Epoch #'], results_epochs['Predicted FL'], label = "Predicted FL Time", 
#          marker = "x", color = 'blue')


plt.scatter(results_epochs['Epoch #'], results_epochs['Measured SL'], label = "Real SL Time", 
         marker = "o", color = 'red')
plt.plot(results_epochs['Epoch #'], results_epochs['Predicted SL'], label = "Simulated SL Time", 
         color = 'red', linestyle = 'dashed')
# plt.scatter(results_epochs['Epoch #'], results_epochs['Predicted SL'], label = "Predicted SL Time", 
#          marker = "x", color = 'red')

plt.scatter(results_epochs['Epoch #'], results_epochs['Measured PSL'], label = "Real PSL Time", 
         marker = "o", color = 'green')
plt.plot(results_epochs['Epoch #'], results_epochs['Predicted PSL'], label = "Simulated PSL Time", 
         color = 'green', linestyle = 'dashed')
# plt.scatter(results_epochs['Epoch #'], results_epochs['Predicted PSL'], label = "Predicted PSL Time", 
#          marker = "x", color = 'green')

plt.scatter(results_epochs['Epoch #'], results_epochs['Measured FSL'], label = "Real FSL Time", 
         marker = "o", color = 'yellow')
plt.plot(results_epochs['Epoch #'], results_epochs['Predicted FSL'], label = "Simulated FSL Time", 
         color = 'yellow', linestyle = 'dashed')
# plt.scatter(results_epochs['Epoch #'], results_epochs['Predicted FSL'], label = "Predicted FSL Time", 
#          marker = "x", color = 'yellow')

# Define your x ticks
x_ticks = results_epochs['Epoch #']

# Define your x and y axis limits
plt.ylim(0, 100)

# Set the legend and show the plot
plt.legend()

# Save your image
file_name = "Epochs_Real_VS_Simulator_V2"
plt.savefig(fname = './Results/Epochs/' + file_name + '.png')

# Show your image
plt.show()
```
