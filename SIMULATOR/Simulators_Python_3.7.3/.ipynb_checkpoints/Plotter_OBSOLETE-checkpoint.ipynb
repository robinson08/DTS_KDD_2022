# Import necessary libraries

```python
#Pandas and plotting tools
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Plotting

```python
# Plot

x_variables = ['NN Flops [GFLOPs]', 'NN Memory [GB]', 'Avg. Processor Power [GFLOP/s]',
               'Avg. Processor Full Residual [GB]', 'Avg. Link Bandwidth [GB/s]', 'Min. Clients',
              "Avg. Input Data Size [GB]", "Avg. Intermediate Results Size [GB]", "Avg. Weights Data Size [GB]"]

x_names = ['NN Flops [GFLOPs]', 'NN Memory [GB]', 'Avg. Processor Power [GFLOP/s]',
           'Avg. Processor Full Residual [GB]', 'Avg. Link Bandwidth [GB/s]', 'Split / Clients', 
          'Input Data [GB]', 'Intermediate Results [GB]', 'Weights Output [GB]']

file_suffixes = ['NN_Size', 'NN_Size', 'Power', 'Memory', 'Bandwidth', 'Split', 'Data_in',
                 'Intermediate', 'Weights']
# NN_Size is repeated because this file is used for both the NN Flops and NN Memory plots

for i in range(len(x_variables)):
    
    ################################# PRE-PROCESSING #################################
    # Read the dataframe
    results_df = pd.read_csv("./Results/VARIABLE_" + file_suffixes[i] + ".csv")
    results_df = results_df.drop(results_df.columns[0], axis=1)
    
    # Remove all "None Feasible" results
    results_filtered = results_df[results_df['Best Architecture'] != "None feasible"]

    # All Runtimes where the architecture Outcome is NOT 'Success' will be turned to None 
    # so they are not plotted later
    results_filtered.loc[results_filtered['PL Outcome'] != 'Success', 'PL Runtime [s]'] = None
    results_filtered.loc[results_filtered['FL Outcome'] != 'Success', 'FL Runtime [s]'] = None
    results_filtered.loc[results_filtered['SL Outcome'] != 'Success', 'SL Runtime'] = None
    results_filtered.loc[results_filtered['PSL Outcome'] != 'Success', 'PSL Runtime [s]'] = None
    results_filtered.loc[results_filtered['FSL Outcome'] != 'Success', 'FSL Runtime [s]'] = None

    # Now get a Dataframe with only the best train times and architectures
    best_train_times = results_filtered[['Best Architecture', 'Best Train Time [s]']]

    # Add new columns to this new dataframe with all values as None
    list_none = [None] * best_train_times.shape[0]
    # Using DataFrame.insert() to add a column at the end of the dataframe
    best_train_times.insert(best_train_times.shape[1], 'Best Train Time PL [s]', list_none, True)
    best_train_times.insert(best_train_times.shape[1], 'Best Train Time SL [s]', list_none, True)
    best_train_times.insert(best_train_times.shape[1], 'Best Train Time FL [s]', list_none, True)
    best_train_times.insert(best_train_times.shape[1], 'Best Train Time PSL [s]', list_none, True)
    best_train_times.insert(best_train_times.shape[1], 'Best Train Time FSL [s]', list_none, True)

    # Now fill out each of these columns with the correct Train Time value only IF they are the best architecture
    best_train_times.loc[best_train_times['Best Architecture'] == 'PL', 
                         'Best Train Time PL [s]'] = best_train_times['Best Train Time [s]']
    best_train_times.loc[best_train_times['Best Architecture'] == 'SL', 
                         'Best Train Time SL [s]'] = best_train_times['Best Train Time [s]']
    best_train_times.loc[best_train_times['Best Architecture'] == 'FL', 
                         'Best Train Time FL [s]'] = best_train_times['Best Train Time [s]']
    best_train_times.loc[best_train_times['Best Architecture'] == 'PSL', 
                         'Best Train Time PSL [s]'] = best_train_times['Best Train Time [s]']
    best_train_times.loc[best_train_times['Best Architecture'] == 'FSL', 
                         'Best Train Time FSL [s]'] = best_train_times['Best Train Time [s]']

    # Now save each of these as individual dataframes / lists to iterate over when plotting them
    # The astype() part is important, otherwise the numbers are interpreted as strings sometimes!
    best_train_times_PL = best_train_times['Best Train Time PL [s]'].astype(float)
    best_train_times_FL = best_train_times['Best Train Time FL [s]'].astype(float)
    best_train_times_SL = best_train_times['Best Train Time SL [s]'].astype(float)
    best_train_times_PSL = best_train_times['Best Train Time PSL [s]'].astype(float)
    best_train_times_FSL = best_train_times['Best Train Time FSL [s]'].astype(float)
    
    ############################################### PLOTTING ###############################################
    
    # Set plot dimensions
    plt.figure(figsize = (15,7.5))
    
    # Set the name of the X axis
    plt.xlabel(x_variables[i], size = 15)
    
    # Set the name of the Y axis
    plt.ylabel("Best Train Time [s]", size = 15)
    
    # Set the title of the plot
    plt.title("Best Train Time VS " + x_variables[i],size = 20)
    
    # Plot each of the algorithms' TRAINING TIMES VS x_variable[i]
    plt.scatter(results_filtered[x_variables[i]], best_train_times_PL, label = "PL", 
             marker = "o", color = 'black')
    plt.scatter(results_filtered[x_variables[i]], best_train_times_FL, label = "FL", 
             marker = "X", color = 'blue')
    plt.scatter(results_filtered[x_variables[i]], best_train_times_SL, label = "SL", 
             marker = "D", color = 'red')
    plt.scatter(results_filtered[x_variables[i]], best_train_times_PSL, label = "PSL", 
             marker = "p", color = 'green')
    plt.scatter(results_filtered[x_variables[i]], best_train_times_FSL, label = "FSL", 
             marker = "s", color = 'yellow')
    
    # Set the legend and show the plot
    plt.legend()
    plt.show()
```
